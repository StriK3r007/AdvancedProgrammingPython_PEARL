### Plot

### Scatter

```python
x = np.array([5, 7, 8, 7, 17, 2, 9, 4, 11, 1, 2, 9])
y = np.array([99, 86, 87, 111, 86, 103, 87, 94, 78, 77, 85, 86])
mpp.scatter(x, y)
mpp.xlabel('x-axis')
mpp.ylabel('y-axis')
mpp.title('Scatter')
```

### Bar

```python
x = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August'])
y = np.array([320, 120, 270, 100, 600, 414, 321, 123])

mpp.xlabel('Months')
mpp.ylabel('Profit')

# graph
mpp.plot(x,y, 'o:g')

# bar chart
mpp.bar(x,y)

# histogram chart
# mpp.hist(y)

mpp.show()
```

### Pie

```python
# Pie Chart
x = np.array([1, 2 ,4])
# y = np.array([1, 1, 1])
y = np.array([0.05, 0.05, 0.05])
mylabel = ["C_Drive", "D_Drive", "E_Dirve"]
mpp.pie(x, y, labels=mylabel)
mpp.show
```

### Linear Regression

```python
from scipy import stats
```

```python
x = np.array([6, 1, 2, 3, 4, 5, 11, 16, 10])
y = np.array([0, 3, 2, 5, 1, 7, 5, 6, 8])
slope, intercept, r, p, std_err=stats.linregress(x, y)
def myfunc(x):
    return slope*x+intercept
mymodel = list(map(myfunc, x))
mpp.scatter(x, y)
mpp.plot(x, mymodel)
mpp.show()
```
